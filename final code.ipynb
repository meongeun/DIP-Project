```python
import cv2
import numpy as np
from PIL import Image
import math
import statistics
import PIL.Image as pilimg
from matplotlib import pyplot as plt
font = cv2.FONT_HERSHEY_COMPLEX
```

```python
background = Image.open("crosswalk9.jpg")
width, height= background.size
area=(0,height/4,width, height)
cropped_img = background.crop(area)
cropped_img.save('cropped_img.jpg')
foreground = Image.open("reccross.png")
```

```python

point_list = []
count = 0

def mouse_callback(event, x, y, flags, param):
    global point_list, count, img_original


    # 마우스 왼쪽 버튼 누를 때마다 좌표를 리스트에 저장
    if event == cv2.EVENT_LBUTTONDOWN:
        print("(%d, %d)" % (x, y))
        point_list.append((x, y))

        print(point_list)
        cv2.circle(img_original, (x, y), 3, (0, 0, 255), -1)



cv2.namedWindow('original')
cv2.setMouseCallback('original', mouse_callback)

# 원본 이미지
img_original = cv2.imread('cropped_img.jpg')
#img_original =cropped_img

while(True):

    cv2.imshow("original", img_original)


    height, weight = img_original.shape[:2]


    if cv2.waitKey(1)&0xFF == 32: # spacebar를 누르면 루프에서 빠져나옵니다.
        break


# 좌표 순서 - 상단왼쪽 끝, 상단오른쪽 끝, 하단왼쪽 끝, 하단오른쪽 끝
pts1 = np.float32([list(point_list[0]),list(point_list[1]),list(point_list[2]),list(point_list[3])])
pts2 = np.float32([[0,0],[weight,0],[0,height],[weight,height]])

print(pts1)
print(pts2)

M = cv2.getPerspectiveTransform(pts1,pts2)

img_result = cv2.warpPerspective(img_original, M, (weight,height))

cv2.imwrite("crosswalk__per.png", img_result)
```

```python
img = cv2.imread("crosswalk__per.png", cv2.IMREAD_GRAYSCALE)
image = pilimg.open("crosswalk__per.png")


img1 = cv2.GaussianBlur(img,(5,5),0)
image1=np.array(image)
a=image1.max()
b=image1.min()
thr=4*(a-b)/5
ret3, thr1 = cv2.threshold(img1, thr, 255, cv2.THRESH_BINARY)
cv2.imshow("threshold v=3/4", thr1)
contours, _ = cv2.findContours(thr1, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

```
